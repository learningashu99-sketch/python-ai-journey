```python
import pandas as pd
# import os

# print(os.getcwd())

df = pd.read_csv(r"F:\python ai journey\projects\ai-ml-projects\Sales_Data_Analysis_Dashboard\data\sales_data.csv")
print("Row, Column: ",df.shape)

# print("\nSome of the dataset values: \n")
# print(df.head())




```

```python
#Checking missing column and duplicate orws
print("\nColumn:- Missing Value:\n",df.isnull().sum())
print("\nDuplicate Rows:\n",df.duplicated().sum())

#Change the object to datetime 

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

print("Dataset info: \n")
df.info()
```

```python
#Group by Category

print("Total sales by each Category:\n")
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)

profit = df.groupby("Category")["Profit"].sum()
print("\nThe profit by each category:\n\n",profit)

average_discount = df.groupby("Category")["Discount"].mean()
print("\nAverage Disocunt by each Categiry:\n\n",average_discount)

```

```python
#Group by Sub-Caetgory

profit_subc = df.groupby("Sub-Category")["Profit"].sum()
print("\nTotal Profit by Sub-Category:\n\n",profit_subc)
```
